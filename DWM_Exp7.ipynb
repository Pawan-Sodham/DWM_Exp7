```python
import pandas as pd
import matplotlib.pyplot as plt
data = {'Values': [15, 33, 48, 52, 68, 72, 81, 85, 92, 99]}
df = pd.DataFrame(data)

bins = [0, 25, 50, 75, 100]
labels = ['Low', 'Medium', 'High', 'Very High']
df['Categories'] = pd.cut(df['Values'], bins=bins, labels=labels)

print(df)
plt.hist(df['Values'], bins=[0, 25, 50, 75, 100], color='blue', edgecolor='black')
plt.title('Value Distribution')
plt.xlabel('Range')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
